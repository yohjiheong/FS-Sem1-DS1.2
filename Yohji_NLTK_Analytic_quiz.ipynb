##### a. Open the text file for processing: First, we are going to open and read the file which we want to analyze eg. The fishing documentation in txt file (page2). Reference: https://huntfish.mdc.mo.gov/sites/default/files/downloads/page/IntroToFishing_2017_v2.pdf

```python
import PyPDF2

with open('fishing.pdf','rb') as pdf_file, open('fishing.txt', 'w', encoding='utf-8') as text_file:
    read_pdf = PyPDF2.PdfFileReader(pdf_file)
    number_of_pages = read_pdf.getNumPages()
    for page_number in range(number_of_pages):
        page = read_pdf.getPage(page_number)
        page_content = page.extractText()
        text_file.write(page_content)
```

```python
page_content
```

##### b. Import required libraries: For various data processing cases in NLP, we need to import some libraries. In this case, we are going to use NLTK for Natural Language Processing. We will use it to perform various operations on the text.

```python
import nltk
```

##### c. Sentence tokenizing: By tokenizing the text with sent_tokenize( ), we can get the text as sentences.

```python
from nltk.tokenize import sent_tokenize, word_tokenize

sentence = open('fishing.txt', encoding='utf-8').read()
sentences = nltk.sent_tokenize(sentence)

print(sentences)
print(len(sentences))
```

##### d. Word tokenizing: By tokenizing the text with word_tokenize( ), we can get the text as words.

```python
words = word_tokenize(sentence)

print(words)
print(len(words))
```

##### e. Find the frequency distribution: Let’s find out the frequency of words in our text.
##### f. Plot the frequency graph: Let’s plot a graph to visualize the word distribution in our text.

```python

from nltk.probability import FreqDist
import matplotlib.pyplot as plt

fdist= FreqDist(words)
print(fdist)

fdist.plot(25,cumulative=False)
plt.show()
```

##### g. Remove punctuation marks: Next, we are going to remove the punctuation marks as they are not very useful for us. We are going to use   isalpha( ) method to separate the punctuation marks from the actual text. Also, we are going to make a new list called words_no_punc, which will store the words in lower case but exclude the punctuation marks.

```python
words_no_punc = [n.lower() for n in words if n.isalpha()]
print (words_no_punc)
print (len(words_no_punc))
```

##### h. Plotting graph without punctuation marks:

```python
frequentwnp= FreqDist(words_no_punc)
print(frequentwnp)

frequentwnp.plot(25,cumulative=False)
plt.show()
```

##### i. List of stopwords:

```python
from nltk.corpus import stopwords
stop_words=set(stopwords.words("english"))
stop_words
```

##### j. Removing stopwords:

```python
filtereds=[]
for wordz in words_no_punc:
    if wordz not in stop_words:
        filtereds.append(wordz)
        
print(filtereds)
```

##### k. Final frequency distribution: the final graph has many useful words that help us understand what our sample data is about, showing how essential it is to perform data cleaning on NLP.

```python
final= FreqDist(filtereds)
print(final)

final.plot(25,cumulative=False)
plt.show()
```
